## Задача 3:
Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалите такие значения из списка и посчитайте суммарные затраты
[100, 125, -90, 345, 655, -1, 0, 200]
Используйте list comprehensions

```python
def clear_data(data : list)->list:
    return [i if i > 0 else 0 for i in data]   

current_costs = clear_data([100, 125, -90, 345, 655, -1, 0, 200])
print(f'Актуальные затраты по периодам {current_costs} и их сумма {sum(current_costs)}')

# Без созданния актуального спиcка расходов

print(f'Текущие расходы {sum([i for i in [100, 125, -90, 345, 655, -1, 0, 200] if i > 0])}')
```

### 4.2 Найдите выручку компании в зависимости от месяца
Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка.
Используйте аннотирование типов.

```python
date = ['2021-09-14', '2021-12-15', '2021-09-08', '2021-12-05', '2021-10-09', '2021-09-30', '2021-12-22', '2021-11-29',
        '2021-12-24', '2021-11-26', '2021-10-27', '2021-12-18', '2021-11-09', '2021-11-23', '2021-09-27', '2021-10-02',
        '2021-12-27', '2021-09-20', '2021-12-13', '2021-11-01', '2021-11-09', '2021-12-06', '2021-12-08', '2021-10-09',
        '2021-10-31', '2021-09-30', '2021-11-09', '2021-12-13', '2021-10-26', '2021-12-09']

purchases = [1270, 8413, 9028, 3703, 5739, 4095, 295, 4944, 5723, 3701, 4471, 651, 7037, 4274, 6275, 4988, 6930, 2971,
            6592, 2004, 2822, 519, 3406, 2732, 5015, 2008, 316, 6333, 5700, 2887]

report_month = 11

def sampling_month(data : list, data2 : list, mounth : str)->list :
    user_date = []
    for i in range(len(data)-1):
        if mounth in data[i]:
            user_date.append(data2[i])
    return user_date

def sampling_month_lc(data : list, data2 : list, mounth : str)->list :    
    return [data2[i] for i in range(len(data)-1) if mounth in data[i]]

print(sum(sampling_month(date,purchases,f'-{report_month}-')))
print(sum(sampling_month_lc(date,purchases,f'-{report_month}-')))

# Решение через lc и ручной ввод месяца

print(sum([purchases[i] for i in range(len(date)-1) if '-11-' in date[i]]))

```
